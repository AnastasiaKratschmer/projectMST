```python
#for cost graph in random
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


file_path = 'testing_outputs/random_strs_tests.csv'

# Read CSV into DataFrame
random_cost = pd.read_csv(file_path)

# Display the DataFrame
print(random_cost)

random_cost = random_cost[random_cost['nr_of_str'] == 10]

#Example: Pivot the DataFrame using melt
random_cost = pd.melt(
    random_cost,
   id_vars=['nr_of_str','str_len'],
    value_vars=['Kruskal', 'Prim', 'Gusfield'],
    var_name='type',
    value_name='cost'
)
print(random_cost)

# Example: Grouped bar chart
plt.figure(figsize=(10, 6))
custom_palette = {"Kruskal": "#a6cee3", "Prim": "#b2df8a", "Gusfield": "#1f78b4"}

sns.barplot(x='str_len', y='cost', hue='type', data=random_cost,palette=custom_palette)

# Set labels and title
plt.xlabel('String Length')
plt.ylabel('Cost')
plt.title('Cost of alignment of 10 strings using Kruskal MST, Prim MST and Gusfield star tree as guide (3 reps)')

# Show the plot
plt.savefig('testing_outputs/graph_random_strs_tests.png')
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'total_cost': {'Kruskal MST': 4745.90, 'Gusfield guide tree': 4348.80},
    'cost_guide_tree': {'Kruskal MST': 1049.65, 'Gusfield guide tree': 1072.70},
}

# Create a multi-index DataFrame
df = pd.DataFrame(data)

# Plotting
fig, ax = plt.subplots(figsize=(8, 6))

# Plot total costs
df['total_cost'].plot(kind='bar', ax=ax, color=['#c2a5cf', '#a6dba0'], label='Total Cost')

# Plot guide tree costs as a portion of the total costs
bottom_values =df['cost_guide_tree']
df['cost_guide_tree'].plot(kind='bar', ax=ax, color=['#7b3294', '#008837'], label='Guide Tree Cost')

# Set labels and title
ax.set_xlabel('Algorithm for guide tree')
ax.set_ylabel('Total Cost Average')
ax.set_title('Total cost vs. guide tree cost, random strings, \nstring length= 100, strings= 6, reps=20')
y_heights = [data['cost_guide_tree']['Gusfield guide tree'],data['total_cost']['Gusfield guide tree']]
for y in y_heights:
    ax.axhline(y=y, color='black', linestyle='--', linewidth=1.2)
# Show the plot
plt.ylim(0, 4900)  # Adjust the y-axis limits as needed
plt.text(-0.12, 500, 'Kruskal \nMST cost', fontsize=12, color='white')
plt.text(-0.15, 2500, 'Total cost of\nalignment', fontsize=12, color='white')
plt.text(0.90, 400, 'Gusfield \nguide \ntree cost', fontsize=12, color='white')
plt.text(0.85, 2500, 'Total cost of\nalignment', fontsize=12, color='white')
plt.tick_params(axis='x', labelrotation=0, pad=10)
plt.savefig('testing_outputs/graph_random_guide_vs_total_cost.png')
plt.show()
```
