```python
#for cost graph in random
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


file_path = 'testing_outputs/random_strs_tests.csv'

# Read CSV into DataFrame
random_cost = pd.read_csv(file_path)

# Display the DataFrame
print(random_cost)

random_cost = random_cost[random_cost['nr_of_str'] == 10]

#Example: Pivot the DataFrame using melt
random_cost = pd.melt(
    random_cost,
   id_vars=['nr_of_str','str_len'],
    value_vars=['Kruskal', 'Prim', 'Gusfield'],
    var_name='type',
    value_name='cost'
)
print(random_cost)

# Example: Grouped bar chart
plt.figure(figsize=(10, 6))
custom_palette = {"Kruskal": "#a6cee3", "Prim": "#b2df8a", "Gusfield": "#1f78b4"}

sns.barplot(x='str_len', y='cost', hue='type', data=random_cost,palette=custom_palette)

# Set labels and title
plt.xlabel('String Length')
plt.ylabel('Cost')
plt.title('Cost of alignment of 10 strings using Kruskal MST, Prim MST and Gusfield star tree as guide (3 reps)')

# Show the plot
plt.savefig('testing_outputs/graph_random_strs_tests.png')
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'total_cost': {'Kruskal MST': 4745.90, 'Gusfield guide tree': 4348.80},
    'cost_guide_tree': {'Kruskal MST': 1049.65, 'Gusfield guide tree': 1072.70},
}

# Create a multi-index DataFrame
df = pd.DataFrame(data)

# Plotting
fig, ax = plt.subplots(figsize=(8, 6))

# Plot total costs
df['total_cost'].plot(kind='bar', ax=ax, color=['#c2a5cf', '#a6dba0'], label='Total Cost')

# Plot guide tree costs as a portion of the total costs
bottom_values =df['cost_guide_tree']
df['cost_guide_tree'].plot(kind='bar', ax=ax, color=['#7b3294', '#008837'], label='Guide Tree Cost')

# Set labels and title
ax.set_xlabel('Algorithm for guide tree')
ax.set_ylabel('Total Cost Average')
ax.set_title('Total cost vs. guide tree cost, random strings, \nstring length= 100, strings= 6, reps=20')
y_heights = [data['cost_guide_tree']['Gusfield guide tree'],data['total_cost']['Gusfield guide tree']]
for y in y_heights:
    ax.axhline(y=y, color='black', linestyle='--', linewidth=1.2)
# Show the plot
plt.ylim(0, 4900)  # Adjust the y-axis limits as needed
plt.text(-0.12, 500, 'Kruskal \nMST cost', fontsize=12, color='white')
plt.text(-0.15, 2500, 'Total cost of\nalignment', fontsize=12, color='white')
plt.text(0.90, 400, 'Gusfield \nguide \ntree cost', fontsize=12, color='white')
plt.text(0.85, 2500, 'Total cost of\nalignment', fontsize=12, color='white')
plt.tick_params(axis='x', labelrotation=0, pad=10)
plt.savefig('testing_outputs/graph_random_guide_vs_total_cost.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Assuming your data is stored in a CSV file named 'data.csv'
data = pd.read_csv('testing_outputs/find_MST_time_new_prim.csv')

data['nr_of_str'] = data['nr_of_str'].astype(int)
data['str_len'] = data['str_len'].astype(int)
data['time_MST,'] = data['time_MST'].astype(float)
data['time_MST_log'] = np.log10(data['time_MST'])
data['strings_log'] = np.log10(data['nr_of_str'])

# Filter data for Kruskal's and Prim's algorithms
kruskal_data = data[data['type'] == 'krus']
prim_data = data[data['type'] == 'prim']



# Plotting
plt.figure(figsize=(10, 6))

# Kruskal's Algorithm
plt.scatter(kruskal_data['strings_log'] + 0.2, kruskal_data['time_MST_log'], label='Kruskal')

# Prim's Algorithm
plt.scatter(prim_data['strings_log'] - 0.2, prim_data['time_MST_log'], label='Prim')

# Linear regression for Kruskal's Algorithm
kruskal_fit = np.polyfit(kruskal_data['strings_log'], kruskal_data['time_MST_log'], 1)
kruskal_trendline = np.poly1d(kruskal_fit)
plt.plot(kruskal_data['strings_log'], kruskal_trendline(kruskal_data['strings_log']), '--', label='Kruskal Trendline')

# Linear regression for Prim's Algorithm
prim_fit = np.polyfit(prim_data['strings_log'], prim_data['time_MST_log'], 1)
prim_trendline = np.poly1d(prim_fit)
plt.plot(prim_data['strings_log'], prim_trendline(prim_data['strings_log']), '--', label='Prim Trendline')

plt.xlabel('Log(Number of Strings)')
plt.ylabel('Log(Time) (seconds)')
plt.title('Comparison of Kruskal\'s and Prim\'s Algorithms with Trendlines')
plt.legend()
plt.show()

# Display the equations of the trendlines
print("Kruskal Trendline Equation: ", kruskal_trendline)
print("Prim Trendline Equation: ", prim_trendline)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Assuming your data is stored in a CSV file named 'your_data.csv'
data = pd.read_csv('testing_outputs/worst_case_vs_actual.csv')

data['ratio_int_ex'] = data['ex_var'] / data['int_var']

# Separate data for Kruskal and Gusfield
kruskal_data = data[data['type'] == 'kruskal']
gusfield_data = data[data['type'] == 'gusfield']

# Plotting
plt.figure(figsize=(10, 6))

# Polynomial regression degree (adjust as needed)
degree = 2

# Kruskal's Algorithm
kruskal_coefficients = np.polyfit(kruskal_data['ratio_int_ex'], kruskal_data['ratio'], degree)
kruskal_poly = np.poly1d(kruskal_coefficients)
kruskal_predictions = kruskal_poly(kruskal_data['ratio_int_ex'])
sns.scatterplot(x='ratio_int_ex', y='ratio', data=kruskal_data, label='Kruskal', marker='o')
sns.lineplot(x=kruskal_data['ratio_int_ex'], y=kruskal_predictions, label=f'Kruskal (Degree {degree})')

# Gusfield's Algorithm
gusfield_coefficients = np.polyfit(gusfield_data['ratio_int_ex'], gusfield_data['ratio'], degree)
gusfield_poly = np.poly1d(gusfield_coefficients)
gusfield_predictions = gusfield_poly(gusfield_data['ratio_int_ex'])
sns.scatterplot(x='ratio_int_ex', y='ratio', data=gusfield_data, label='Gusfield', marker='x')
sns.lineplot(x=gusfield_data['ratio_int_ex'], y=gusfield_predictions, label=f'Gusfield (Degree {degree})')

plt.xlabel('Ratio d_e/d_i')
plt.ylabel('Total cost/worst case cost')
plt.title('Comparison of Kruskal\'s and Gusfield\'s Algorithms')
plt.legend()

# Display the formulas
print(f'Kruskal Formula: {np.poly1d(kruskal_coefficients)}')
print(f'Gusfield Formula: {np.poly1d(gusfield_coefficients)}')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Assuming your data is stored in a CSV file named 'your_data.csv'
# (Assuming you have already imported pandas as pd)
data = pd.read_csv('testing_outputs/worst_case_vs_actual_new_distance_k8_c2.csv')
# Convert 'int_var' and 'ex_var' to floats
data['int_var'] = data['int_var'].astype(float)
data['ex_var'] = data['ex_var'].astype(float)

# Separate data for Kruskal and Gusfield
kruskal_data = data[data['type'] == 'kruskal']
gusfield_data = data[data['type'] == 'gusfield']

# Combine all data to find the global min and max values
combined_data = pd.concat([kruskal_data, gusfield_data])
global_min = combined_data['ratio'].min()
global_max = combined_data['ratio'].max()

# Create a contour map for Kruskal
plt.figure(figsize=(12, 8))
plt.subplot(2, 1, 1)
kruskal_heatmap = kruskal_data.pivot_table(index='ex_var', columns='int_var', values='ratio', aggfunc='mean')
x = kruskal_heatmap.columns.astype(float)
y = kruskal_heatmap.index.astype(float)
X, Y = np.meshgrid(x, y)
Z = kruskal_heatmap.to_numpy()

contour = plt.contourf(X, Y, Z, cmap='coolwarm', levels=20, vmin=0.4, vmax=0.8)
plt.colorbar(contour, label='Ratio', ticks=np.linspace(global_min, global_max, 11))
plt.title('Kruskal: (len/len worst case), 8 nodes, 2 clusters')
plt.xlabel('d_i')
plt.ylabel('d_e')

# Create a contour map for Gusfield
plt.subplot(2, 1, 2)
gusfield_heatmap = gusfield_data.pivot_table(index='ex_var', columns='int_var', values='ratio', aggfunc='mean')
x = gusfield_heatmap.columns.astype(float)
y = gusfield_heatmap.index.astype(float)
X, Y = np.meshgrid(x, y)
Z = gusfield_heatmap.to_numpy()

contour = plt.contourf(X, Y, Z, cmap='coolwarm', levels=20, vmin=0.4, vmax=0.8)
plt.colorbar(contour, label='Ratio', ticks=np.linspace(global_min, global_max, 11))
plt.title('Gusfield: (len/len worst case), 8 nodes, 2 clusters')
plt.xlabel('d_i')
plt.ylabel('d_e')

plt.tight_layout()
plt.savefig('testing_outputs/contour_ratio_worst_case_real_8k,2c')
plt.show()
```
