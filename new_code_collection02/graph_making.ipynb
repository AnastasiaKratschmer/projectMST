```python
#for cost graph in random
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


file_path = 'testing_outputs/random_strs_tests.csv'

# Read CSV into DataFrame
random_cost = pd.read_csv(file_path)

# Display the DataFrame
print(random_cost)

random_cost = random_cost[random_cost['nr_of_str'] == 10]

#Example: Pivot the DataFrame using melt
random_cost = pd.melt(
    random_cost,
   id_vars=['nr_of_str','str_len'],
    value_vars=['Kruskal', 'Prim', 'Gusfield'],
    var_name='type',
    value_name='cost'
)
print(random_cost)

# Example: Grouped bar chart
plt.figure(figsize=(10, 6))
custom_palette = {"Kruskal": "#a6cee3", "Prim": "#b2df8a", "Gusfield": "#1f78b4"}

sns.barplot(x='str_len', y='cost', hue='type', data=random_cost,palette=custom_palette)

# Set labels and title
plt.xlabel('String Length')
plt.ylabel('Cost')
plt.title('Cost of alignment of 10 strings using Kruskal MST, Prim MST and Gusfield star tree as guide (3 reps)')

# Show the plot
plt.savefig('testing_outputs/graph_random_strs_tests.png')
plt.show()

```

```python
#for cost graph in family 1 (number of fams vs. str pr fam)
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


file_path = 'testing_outputs/familiy_outputs_testing.csv'

# Read CSV into DataFrame
fam_cost = pd.read_csv(file_path)

# Display the DataFrame
print(fam_cost)

#fam_cost1 = fam_cost[fam_cost['str_len'] == 200]
fam_cost1=fam_cost[fam_cost['internal_var'] == 0.1]
fam_cost1=fam_cost1[fam_cost1['external_var'] == 0.9]

#Example: Pivot the DataFrame using melt
fam_cost1 = pd.melt(
    fam_cost1,
   id_vars=['nr_of_fams','str_pr_fam','str_len'],
    value_vars=['Kruskal', 'Prim', 'Gusfield'],
    var_name='type',
    value_name='cost'
)
print(fam_cost1)

# Example: Grouped bar chart
plt.figure(figsize=(10, 6))
custom_palette = {"Kruskal": "#a6cee3", "Prim": "#b2df8a", "Gusfield": "#1f78b4"}

g = sns.FacetGrid(fam_cost1, col='str_pr_fam', row='nr_of_fams', margin_titles=True)
g.map(sns.barplot, 'str_len', 'cost', 'type', palette=custom_palette)

# Set labels and title
plt.xlabel('String Length')
plt.ylabel('Cost')
plt.title('Cost of alignment of 10 strings using Kruskal MST, Prim MST and Gusfield star tree as guide (3 reps)')

# Show the plot
plt.savefig('testing_outputs/fam1.png')
plt.show()



```

```python
#for cost graph in family 2 (internal vs. external var)
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


file_path = 'testing_outputs/familiy_outputs_testing.csv'

# Read CSV into DataFrame
fam_cost = pd.read_csv(file_path)

# Display the DataFrame
print(fam_cost)

#fam_cost2 = fam_cost[fam_cost['str_len'] == 200]
fam_cost2=fam_cost[fam_cost['nr_of_fams'] == 4]
fam_cost2=fam_cost2[fam_cost2['str_pr_fam'] == 5]
fam_cost2=fam_cost2[fam_cost2['external_var'] != 0.5]
fam_cost2=fam_cost2[fam_cost2['internal_var'] != 0.5]

#Example: Pivot the DataFrame using melt
fam_cost2 = pd.melt(
    fam_cost2,
   id_vars=['str_len','internal_var','external_var'],
    value_vars=['Kruskal', 'Prim', 'Gusfield'],
    var_name='type',
    value_name='cost'
)
print(fam_cost2)

# Example: Grouped bar chart
plt.figure(figsize=(10, 6))
custom_palette = {"Kruskal": "#a6cee3", "Prim": "#b2df8a", "Gusfield": "#1f78b4"}

g = sns.FacetGrid(fam_cost2, col='internal_var', row='external_var', margin_titles=True)
g.map(sns.barplot, 'str_len', 'cost', 'type', palette=custom_palette)

# Set labels and title
plt.xlabel('String Length')
plt.ylabel('Cost')


# Show the plot
plt.savefig('testing_outputs/fam2.png')
plt.show()



```
