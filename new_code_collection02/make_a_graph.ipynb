```python
import matplotlib.pyplot as plt
import pandas as pd
from io import StringIO

data_str = """5,733.07,730,736.27,736.27,733.07,731.8
10,1643.07,1651.73,1674.2,1661.53,1643.07,1681.07
50,7971.93,7972.6,7977.67,8010.07,7971.93,7999
100,15984.47,16022.6,15943.47,15990.53,15984.47,16078.93
200,32239.2,32328.67,32339.07,32264,32239.2,32199.6"""

# Parse CSV data into a DataFrame
df = pd.read_csv(StringIO(data_str), names=['str_len', 'zero', 'mid', 'ex', 'rd', 'min_sn', 'max_sn'])

# Plotting
bar_width = 0.15
index = np.arange(len(df))  # x-axis values

for i, metric in enumerate(df.columns[1:]):
    plt.bar(index + i * bar_width, df[metric], bar_width, label=metric)

# Customize plot
plt.xlabel('String Length')
plt.ylabel('Values')
plt.title('Cost of alignment 10 strings (15 reps) for starting node strategy')
plt.xticks(index + bar_width * ((len(df.columns) - 1) / 2), df['str_len'])
plt.legend()

# Show the plot
plt.show()
```
