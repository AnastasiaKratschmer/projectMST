```python
from functions_collection_copy import *
import time
import statistics
import csv
import random as random

score_matrix={'a': {'a': 0, 'c': 5, 'g': 2, 't': 5}, 'c': {'a': 5, 'c': 0, 'g': 5, 't': 2}, 'g': {'a': 2, 'c': 5, 'g': 0, 't': 5}, 't': {'a': 5, 'c': 2, 'g': 5, 't': 0}}
gap_cost=5
seqs=['aaatt',
 'cagtcgtt',
 'agctg',
 'ctgtagaac',
 'gcagg',
 'tttttt',
 'agcggcg',
 'cacccgtga',
 'tctgcag',
 'taactcata']
```

```python
csv_file_path='short_string_times_with_prim'
str_lens=[5,10,20]
for nr in str_lens:
    gradual_time=[]
    OBO_time=[]
    Gus_time=[]
    Prim_time=[]
    repeats=10
    nr_of_str=10
    len_str=nr
    random_sequences_collection=[]
    for i in range(0,repeats):
        random_pile=[]
        for j in range (0,nr_of_str):
            a_sequence=[]
            for k in range(0,len_str):
                a_sequence.append(random.choice(['a','c','t','g']))
            a_sequence=''.join(a_sequence)
            random_pile.append(a_sequence)
        random_sequences_collection.append(random_pile)

    for random_pile in random_sequences_collection:
        start_time=time.time()
        new_assembly_gradual_x(random_pile,score_matrix,gap_cost)
        end_time=time.time()
        execution_time= end_time-start_time
        gradual_time.append(execution_time)


        start_time=time.time()
        new_assembly_OBO_x(random_pile,score_matrix,gap_cost)
        end_time=time.time()
        execution_time= end_time-start_time
        OBO_time.append(execution_time)


        start_time=time.time()
        new_assembly_Gus_x(random_pile,score_matrix,gap_cost)
        end_time=time.time()
        execution_time= end_time-start_time
        Gus_time.append(execution_time)

        start_time=time.time()
        new_assembly_Prim_x(random_pile,score_matrix,gap_cost)
        end_time=time.time()
        execution_time= end_time-start_time
        Prim_time.append(execution_time)

    print("Mean gradual with str len ",str(nr),": ", str(statistics.mean(gradual_time)))
    print("Mean OBO with str len ",str(nr),": ", str(statistics.mean(OBO_time)))
    print("Mean Gus with str len ",str(nr),": ", str(statistics.mean(Gus_time)))
    print("Mean Prim with str len ",str(nr),": ", str(statistics.mean(Prim_time)))
    with open(csv_file_path, 'a', newline='') as csvfile:
                            csvwriter = csv.writer(csvfile)
                            csvwriter.writerow([nr,statistics.mean(gradual_time),statistics.mean(OBO_time),statistics.mean(Gus_time),statistics.mean(Prim_time)])
```
